# Iris

```python
iris = load_iris_to_df()
iris_label = iris.get('label')
iris = iris.drop('label', 1)
pca_iris = reduce_dimension(iris, 'PCA', 2, plot_curve=True)
ica_iris = reduce_dimension(iris, 'ICA', 2)
rnd_iris = reduce_dimension(iris, 'random', 2)
```

```python
plot_2d(pca_iris.assign(label=iris_label), "0", "1")
plot_2d(ica_iris.assign(label=iris_label), "0", "1")
plot_2d(rnd_iris.assign(label=iris_label), "0", "1")
```

```python
kmeans_pca_iris = cluestering(pca_iris, 'kmeans', 3)
em_pca_iris = cluestering(pca_iris, 'EM', 3)
plot_2d(kmeans_pca_iris, "0", "1")
plot_2d(em_pca_iris, "0", "1")
```

```python
kmeans_ica_iris = cluestering(ica_iris, 'kmeans', 3)
em_ica_iris = cluestering(ica_iris, 'EM', 3)
plot_2d(kmeans_ica_iris, "0", "1")
plot_2d(em_ica_iris, "0", "1")
```

```python
kmeans_rnd_iris = cluestering(rnd_iris, 'kmeans', 3)
em_rnd_iris = cluestering(rnd_iris, 'EM', 3)
plot_2d(kmeans_rnd_iris, "0", "1")
plot_2d(em_rnd_iris, "0", "1")
```
